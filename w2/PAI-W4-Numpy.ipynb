```python
## Tính điểm thi trung bình cho từng bạn sinh viên

def averge(x):
    return (x[:, 9] - x[:, 5])/3
# Cột ngoài cùng là điểm trung bình cộng cả 3 môn đã được thêm vào
data = np.hstack((data, np.array([averge(data)]).T))
data
```

```python
## Tính điểm thi trung bình theo từng môn cho toàn bộ sinh viên
def subject(x,idx):
    return sum(x[:, idx])/len(x)

print(f'Điểm trung bình môn Văn: {subject(data, 6)}')
print(f'Điểm trung bình môn Anh: {subject(data, 7)}')
print(f'Điểm trung bình môn Toán: {subject(data, 8)}')
```

```python
## Đưa ra điểm thi trung bình cao nhất và thấp nhất cho toàn bộ 3 môn.
print(f'Điểm thi trung bình cao nhất {np.max(data[: , -1])}')
print(f'Điểm thi trung bình thấp nhất {np.min(data[: , -1])}')
```

```python
data[:, :-1]
```

```python
## Xếp danh sách dựa vào điểm trung bình của môn Toán + Anh, tìm bạn có vị trí 100
def sortby_Math_Eng(x):
    lis = np.array([x[:, 7] + x[:, 8]]) / 2
    x = np.hstack((x, lis.T))
    x = x[np.argsort(x[:, -1])]
    x = x[::-1]
    return x[:, :-1]
```

```python
# Danh sách sau khi đã sort
sortby_Math_Eng(data)
```

```python
# Đứa loi nhoi thứ 100
sortby_Math_Eng(data)[99]
```

```python
## c) Điểm trung bình 3 môn từ cao xuống thấp
def sortby_c(x):
    x = x[np.argsort(x[:, -1])]
    x = x[::-1]
    return x[:299]
```

```python
sortby_c(data)[:10] # Cột điểm trung bình cộng là cột ngoài cùng
```

```python
## So sánh 2 bạn sinh viên mã số: 60575 và 60602 dựa trên tổng điểm 3 môn (không tính điểm cộng)
def compare(x, y):
    if sum(x[6:9]) > sum(y[6:9]):
        return f'{x[0]} có điểm cao hơn {y[0]}'
    elif sum(x[6:9]) < sum(y[6:9]):
        return f'{y[0]} có điểm cao hơn {x[0]}'
    else:
        return f'2 bạn có điểm bằng nhau'

for idx, i in enumerate(data[:, 1]):
    if i == 60575:
        x = data[idx]
    if i == 60602:
        y = data[idx]
        
compare(x, y)
```
